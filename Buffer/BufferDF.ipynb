```python
import geopandas
```

Add two layers and reproject it (meters not degrees)

```python
fires = geopandas.read_file("../Data/Facilities/DemoCairo.gpkg", layer='fire').to_crs("EPSG:3857")
facilities = geopandas.read_file("../Data/Facilities/DemoCairo.gpkg", layer='facilities').to_crs("EPSG:3857")
facilities.head()
```

- type(fires['geometry']) this is a GeoSeries type
- type(fires) is a Geodataframe

```python
fires['geometry'] = fires['geometry'].buffer(distance=500) #500 meter buffer
```

```python
fires.explore()
```

```python
result = geopandas.sjoin(fires, facilities, how='left')
result.to_file('../Data/Facilities/DemoCairoResults.gpkg', driver='GPKG', layer='Result1')
```
